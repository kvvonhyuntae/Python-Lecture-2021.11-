# matplotlib - plot 기본

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family']= 'Malgun Gothic'

```

```python
plt.plot([0, 1, 2, 3],[0, 1, 4, 9]) # x값, y값 리스트
plt.title('2021 부동산 상승률')              #제목
plt.grid()                          #격자 라인
plt.show()                          #합쳐서 화면에 보이게함
```

```python
plt.plot([0, 1, 2, 3],[0, 1, 4, 9], 'ro-') #color, maker, linestyle
plt.title('Red-Circle-Line')
plt.grid()
plt.show()

#color: blue, green, red, cyan, magenta, yellow, black(k), white
#marker: circle(o), 삼각형(^), 역삼각형(v), 사각형(s), 별(*), 플러스(+)
#linestyle: 직선(-), dashed(--), dash-dot(-.), dotted(:)
```

```python
plt.plot([0, 1, 2, 3],[0, 1, 4, 9], 'ms--')
plt.title('Magenta-Square-Dashed')
plt.grid()
plt.show()
```

```python
plt.plot([0, 1, 2, 3],[0, 1, 4, 9], 'ms--')
plt.plot([8, 4, 2, 1], 'g^:')
plt.title('Magenta-Square-Dashed / Green-삼각형- 점선')
plt.grid()
plt.show()
```

- Line Style


```python
plt.plot([0, 1, 2, 3],[0, 1, 4, 9], 'ms--', linewidth=3, markersize=10)
plt.plot([8, 4, 2, 1], 'g^:',lw=5, ms=20)
plt.title('Line Style')
plt.grid()
plt.show()

#사이즈 변켱
```

그림 범위 지정

```python
plt.plot([0, 1, 2, 3],[0, 1, 4, 9], 'ro-') #color, maker, linestyle
plt.title('X축 Y축 범위 지정')
plt.xlim(-1, 5)
plt.ylim(-1, 10)
plt.grid()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 500)
y = np.cos(x)
plt.plot(x, y)
plt.title('Tick 설정')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.show()
```

- 범례(legend)

```python
x = np.linspace(-np.pi, np.pi, 500)
c,s = np.cos(x), np.sin(x)
plt.title('LEgend를 표시한 플롯')
plt.plot(x, c, label='cosine')
plt.plot(x, s, label = 'sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.show()
```

- X 축 Y축 label

```python
x = np.linspace(-np.pi, np.pi, 500)
c,s = np.cos(x), np.sin(x)
plt.title('X축, Y축 라벨을 표시한 플롯')
plt.plot(x, c, label='cosine')
plt.plot(x, s, label = 'sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid(), plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

```python
plt.figure(figsize=(10,6)) #크기설정
plt.title('그림크기지정')
plt.plot(x, c, label='cosine')
plt.plot(x, s, label = 'sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid(), plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('그림크기지정한 그래프를 저장')
plt.plot(x, c, label='cosine')
plt.plot(x, s, label = 'sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid(), plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.savefig('삼각함수.png')  #파일로 저장하는법
```
