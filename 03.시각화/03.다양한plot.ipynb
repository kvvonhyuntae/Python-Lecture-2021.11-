# matplotlib - 다양한 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family']= 'Malgun Gothic'
```

### 1. 산점도(Scatter Plot)

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ans.groupby('dataset').agg(['mean','std'])
```

```python
x, y = np.random.rand(100), np.random.randn(100)
plt.title('산점도')
plt.scatter(x, y)
plt.show()
```

```python
colors = np.random.rand(100)
area = np.pi * (15 * np.random.rand(100)*2)
plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.title('Bubble Chart')
plt.show()
```

```python
# 4. Box plot
```

```python
# Setosa 품종
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length, setosa.petal_width],
            labels=iris.columns[:-1])
plt.title('Setosa 품종의 각 피쳐별 박스 플롯', fontsize=15)
plt.show()
```

### 5. 파이(pie) 차트

```python
ratio = [34, 32, 16, 18]
labels = ['사과', '바나나', '멜론', '포도']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()

#shadow 그림자 넣으라는 말임
```
