# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family']= 'Malgun Gothic'
```

```python
np.random.seed(2021)
df=pd.DataFrame(np.random.randn(100, 3),
                index=pd.date_range('1/1/2021', periods=100),
                columns=list('ABC')).cumsum()
df.plot()
plt.title('pandas의 plo method의 사용예')
plt.xlabel('일자'),plt.ylabel('데이터')
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar', rot=0)
plt.title('Setosa 품종의 sepal length')
plt.show()

#, rot=0로 밑의 숫자의 각도를 조절한다.
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'Feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 feature  별 평균')
plt.ylim(0,10) #y에 리미트를 준다는뜻
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%2.f%%') #글자에 %퍼센트를 포함하겠다는뜻
plt.title('선실별 승객수')
plt.show()
```

## 4. box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize=(8,10))
#박스 플롯에서 사이즈를 조절할때는 plot.box()안에 figsize=(숫자,숫자)를 넣는다.
plt.title('Setosa 품종의 각 Feature별 Box plot')
plt.show()
```

```python
iris.boxplot(by='species')
plt.title('각 피쳐의 품종별 데이터에 대한 box plot')
plt.show()
```
