# 데이터 시각화 - seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family']= 'Malgun Gothic'
```

## 1. Count plot

```python
titanic = sns .load_dataset('titanic')
sns.countplot(x='class', data = titanic)
plt.title('타인타닉호의 각 클래스별 승객 수')
plt.show()
```

```python
sns.countplot(x='embark_town', data=titanic);
```

### 2.pair plot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.title('Iris Data의 pair plot')
plt.show()
```

```python
### 3. Heatmap
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.shape
```

```python
passengers = flights.pivot('month', 'year', 'passengers') #행, 열 데이터
passengers.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(passengers, annot=True, fmt='d', linewidths=1)
plt.show()
```

## 4. box plot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title('요일별/성별 전체요금')
plt.show()
```

```python
sns.boxplot(data=iris[iris.species=='setosa']);
```
